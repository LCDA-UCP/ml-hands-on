```python
# load the wines dataset (../data/wines.csv)
import pandas as pd

df=pd.read_csv("../data/wines.csv")

```

```python

df.head(5)
```

```python
# check for missing values in the dataset

df.isnull().sum()
```

```python
# if any, impute the missing values with the mean of the column

df.fillna(df.mean(), inplace=True)
df.isnull().sum()
```

```python
# scale the dataset using two different scaling methods
#   #######StandardScaler##########
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df_Standard = scaler.fit_transform(df)
df_Standard = pd.DataFrame(df_Standard, columns=df.columns)
df_Standard.head()

```

```python
df_Standard.describe()
```

```python
# ########MinMaxScaler########
from sklearn.preprocessing import MinMaxScaler

scalerMinMax = MinMaxScaler()
df_MinMax =scalerMinMax.fit_transform(df)
df_MinMax = pd.DataFrame(df_MinMax, columns=df.columns)
df_MinMax.head()

```

```python
df_MinMax.describe()
```

```python
# perform feature selection using two different methods
```

```python
##### KBest ######
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import f_regression

#Separating Features and Target
X = df.drop('target', axis=1)
Y = df['target']

KBest= SelectKBest(score_func=f_regression, k=4)
data = KBest.fit_transform(X,Y)
#features names
featurenamesK=X.columns.values[KBest.get_support()]
data = pd.DataFrame(data, columns=featurenamesK)
data
```

```python
# #####Variance Threshold
from sklearn.feature_selection import VarianceThreshold

X = df.drop('target', axis=1)

varThresh = VarianceThreshold(threshold=0.5)
X_var= varThresh.fit_transform(X)
#features names
featurenamesV = X.columns.values[varThresh.get_support()]
dataVar=pd.DataFrame(X_var,columns= featurenamesV)
dataVar
```

```python
# compare the results of the two feature selection methods
#treinar em modelos e ver as accuracy
#Explicar apenas as features que foram retiradas estatiscamente e porque foram retiradas
```

```python
#Para comparar os resultados dos dois metodos de feature selection, vou treinar um modelo com as features selecionadas pelo dois metódos e ver a accuracy de cada um.

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

#Dividir em dados de treino e dados de teste
X_trainK, X_testK, y_train, y_test = train_test_split(X[featurenamesK],Y,test_size=0.2,random_state=42)
X_trainV, X_testV, _,_ =train_test_split(X[featurenamesV],Y,test_size=0.2,random_state=42)

#Treinar o modelo
model_K = RandomForestClassifier(random_state=42)
model_K.fit(X_trainK,y_train)
y_predK= model_K.predict(X_testK)
accuracyK= accuracy_score(y_test,y_predK)

model_V = RandomForestClassifier(random_state=42)
model_V.fit(X_trainV,y_train)
y_predV= model_V.predict(X_testV)
accuracyV = accuracy_score(y_test,y_predV)

print("Accuracy com Kbest é:", accuracyK)
print("Accuracy com varianceThreshold é de:", accuracyV)

```

 Com base nesse resultado podemos ver que o modelo usando as feature do metódo varianceThreshold  teve uma melhor perfomace , tendo a accuracy de 100%.
